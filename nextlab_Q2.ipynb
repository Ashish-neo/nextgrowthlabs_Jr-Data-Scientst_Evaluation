```python
import numpy as np
import pandas as pd
import language_tool_python
tool = language_tool_python.LanguageTool('en-US')  # language English-US
from textblob import TextBlob
```

```python
data=pd.read_csv('/content/review_data.csv',nrows=1000)
pd.set_option("display.max_column", None)
display(data.head())
```

```python
data.shape
```

```python
# check missing values
data.isnull().sum()
```

```python
df=data[['text']]
```

```python
df.head()
```

```python
#check the sentance is correct or not
try:
    df = df.assign(Grammatical_Status = ['Incorrect' if len(tool.check(rows))>0 else 'Correct' for rows in df.text])
except Exception as e:
    print('Error: ', e)
```

```python
df.head()
```

```python
#correct the sentance
try:
    df = df.assign(Corrected = [tool.correct(rows) for rows in df.text])
except Exception as e:
    print('Error: ', e)
```
